```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Pharma_data.csv'  
data = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Region', y='Effectiveness', hue='Product_Name')
plt.title('Average Effectiveness by Drug and Region')
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.legend(title='Product Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Product_Name', y='Effectiveness', inner='box')
plt.title('Distribution of Effectiveness for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Trial_Period', y='Effectiveness')
plt.title('Effectiveness Across Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='Marketing_Spend', y='Effectiveness', scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Effect of Marketing Spend on Effectiveness')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.tight_layout()
plt.show()
```
